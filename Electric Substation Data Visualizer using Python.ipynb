```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
```

```python
#read csv file
jam69 = pd.read_csv('JAM-6920211006093119.CSV', na_values=0)
print (jam69.head(3))
```

```python
#show number of rows and columns
print(jam69.shape)
```

```python
#show column names
print(jam69.columns)
```

```python
#show data types of columns
print (jam69.info())
```

```python
#describe data frame
print (jam69.describe())
```

```python
#convert DATE as a string to Datetime format
jam69["Date_Time"] = jam69["DATE"] + " " + jam69["TIME"]
jam69["Date_Time"] = pd.to_datetime(jam69["Date_Time"])

jam69["HOURLY"] = jam69["Date_Time"].dt.hour

jam69["DATE"] = pd.to_datetime(jam69["DATE"])
jam69["TIME"] = pd.to_datetime(jam69["TIME"]).dt.time


jam69 = jam69[['Date_Time','DATE', 'TIME', 'HOURLY', 'CAL_JAM_PTS,69-Iab Value', 'CAL_JAM_PTS,69-Ibc Value',
       'CAL_JAM_PTS,69-Ica Value', 'CAL_JAM_PTS,69-Vab Value',
       'CAL_JAM_PTS,69-Vbc Value', 'CAL_JAM_PTS,69-Vca Value',
       'CAL_JAM_PTS,Freq Value', 'CAL_JAM_PTS,69-S Value',
       'CAL_JAM_PTS,69-P Value', 'CAL_JAM_PTS,69-Q Value',
       'CAL_JAM_PTS,69-pf Value']]



jam69.info()
jam69.head()
```

```python
#find null values
print(jam69.shape, "\n\n")

print(jam69.isnull().sum())
```

```python
#drop null values
jam69.dropna(inplace=True)
print(jam69.isna().sum(), "\n")

print(jam69.shape)
```

```python
#group data hourly and get the min, max, and mean
jam_hr_min=jam69.groupby("HOURLY")[jam69.columns].min()
jam_hr_max=jam69.groupby("HOURLY")[jam69.columns].max()
jam_hr_mean=jam69.groupby("HOURLY")[jam69.columns].mean()

#verify duplicated data
jam_hr_min_dup = jam_hr_min.duplicated().sum()
jam_hr_mean_dup = jam_hr_mean.duplicated().sum()
jam_hr_max_dup = jam_hr_max.duplicated().sum()

(jam_hr_min_dup, jam_hr_mean_dup, jam_hr_max_dup)
```

```python
#reset index
jam69.reset_index(drop=True, inplace=True)
jam69.head(3)
```

```python
#find Iab, Ibc and Ica its min, max, and mean values
Ill_Iab = jam69['CAL_JAM_PTS,69-Iab Value']
Ill_Ibc = jam69['CAL_JAM_PTS,69-Ibc Value']
Ill_Ica = jam69['CAL_JAM_PTS,69-Ica Value']


Iab_max = Ill_Iab.max()
Ibc_max = Ill_Ibc.max()
Ica_max = Ill_Ica.max()

Iab_mean = round(Ill_Iab.mean(), 2)
Ibc_mean = round(Ill_Ibc.mean(), 2)
Ica_mean = round(Ill_Ica.mean(), 2)

Iab_min = Ill_Iab.min()
Ibc_min = Ill_Ibc.min()
Ica_min = Ill_Ica.min()


Imin = (Iab_min,Ibc_min,Ica_min)
Imean = (Iab_mean, Ibc_mean, Ica_mean)
Imax = (Iab_max,Ibc_max,Ica_max)

Imin,Imean,Imax
```

```python
#plot the values using bar graph

def addlabels(x_i,y_i):
    for i in range(len(x_i)):
        plt.text(x_i[i], y_i[i], y_i[i], ha = 'center', va = 'bottom')

x_i = ["Iab Max", "Iab Mean", "Iab Min", "Ibc Max","Ibc Mean", "Ibc Min", "Ica Max", "Ica Mean", "Ica Min"]
y_i = [Iab_max, Iab_mean, Iab_min, Ibc_max, Ibc_mean, Ibc_min, Ica_max, Ica_mean, Ica_min]

fig, ax = plt.subplots(figsize=(10,5))
ax.bar("Iab Max", Iab_max)
ax.bar("Iab Mean", Iab_mean)
ax.bar("Iab Min", Iab_min)

ax.bar("Ibc Max", Ibc_max)
ax.bar("Ibc Mean", Ibc_mean)
ax.bar("Ibc Min", Ibc_min)

ax.bar("Ica Max", Ica_max)
ax.bar("Ica Mean", Ica_mean)
ax.bar("Ica Min", Ica_min)

addlabels (x_i,y_i)

ax.set_title("Jampason S/s - August 2021")
ax.set_ylabel("Load Current (Amp)")

plt.show()
```

```python
#visualize the hourly load current profile of the substation using line graph

fig, ax = plt.subplots(1, 3, figsize=(15,5), sharex=True, sharey=True)
fig.tight_layout(pad = 5.0)


jam69.groupby("HOURLY")[['CAL_JAM_PTS,69-Iab Value', 'CAL_JAM_PTS,69-Ibc Value', 'CAL_JAM_PTS,69-Ica Value']].max().plot(ax=ax[0])
ax[0].set_ylabel("Load Current (Amp)", fontsize=12)
ax[0].set_xlabel("Hourly", fontsize=12)
ax[0].set_title("Max Values", fontsize=12)
ax[0].legend(["Iab","Ibc","Ica"],loc="upper left")

jam69.groupby("HOURLY")[['CAL_JAM_PTS,69-Iab Value', 'CAL_JAM_PTS,69-Ibc Value', 'CAL_JAM_PTS,69-Ica Value']].mean().plot(ax=ax[1])
ax[1].set_xlabel("Hourly", fontsize=12)
ax[1].set_title("Average Values", fontsize=12)
ax[1].legend(["Iab","Ibc","Ica"],loc="upper left")

jam69.groupby("HOURLY")[['CAL_JAM_PTS,69-Iab Value', 'CAL_JAM_PTS,69-Ibc Value', 'CAL_JAM_PTS,69-Ica Value']].min().plot(ax=ax[2])
ax[2].set_xlabel("Hourly", fontsize=12)
ax[2].set_title("Min Values", fontsize=12)
ax[2].legend(["Iab","Ibc","Ica"],loc="upper left")

plt.show()
```
